```python
def tojson(file):
    dataCSV = pd.read_csv(file)
    series = []
    for i in dataCSV.state.unique():
        subset = dataCSV[dataCSV.state == i].sort_values("date", ascending=True)

        dic = {}
        dic["name"] = i
        for col in subset.columns[2:]:
            dic[col] = list(subset[col])

        series.append(dic)
    
    data = {"series": series, "date": [str(i) for i in list(dataCSV["date"])]}
    with open('states.json', 'w') as outfile:
        json.dump(data, outfile, indent = 4)
```
